```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Configurando o pandas para mostrar 500 linhas na exibição do dataframe
pd.set_option('display.max_rows', 500)
```

```python
df_produto = pd.read_csv("/content/drive/MyDrive/Estudos/Desbravando Pandas/data/produto.csv")
```

# Continuando algumas aplicações do .apply()

```python
# Buscando todos os valores unicos da coluna descItem utilizando o .unique()
df_produto['descItem'].unique()
```

```python
# Criando uma função que verifica se a string passada começa com 'massa'
def is_massa(x):
  return x.lower().startswith('massa')

is_massa('Massa tradicional')
```

```python
# Aplicando a função is_massa para a série da coluna descItem e atribuindo a filto_massa
filtro_massa = df_produto['descItem'].apply(is_massa)
filtro_massa
```

```python
# Usando o filtro_massa no dataframe
df_produto[filtro_massa]
```

```python
# Reescrevendo a mesma função is_massa utilizando lambda function
filtro_massa = df_produto['descItem'].apply(lambda x: x.lower().startswith('massa'))
df_produto[filtro_massa]
```

```python
# Reescrevendo a mesma função is_massa utilizando .str acessando os métodos do tipo de objeto da série descItem
filtro_massa = df_produto['descItem'].str.lower().str.startswith('massa')
df_produto[filtro_massa]
```

# Ordenação

```python
# Ordenando os valores de vlPreco do menor para o maior com .sort_values()
df_produto.sort_values(by='vlPreco')
```

```python
# Ordenando os valores de vlPreco do maior para o menor com .sort_values()
df_produto.sort_values(by='vlPreco', ascending=False)
```

# Novos Filtros

```python
# Criando uma nova coluna contendo os primeiro valor de nome da coluna descItem
df_produto['primeiroNome'] = df_produto['descItem'].apply(lambda x: x.lower().split(" ")[0])
```

```python
# Deletando as linhas que contenham algum valor duplicado com base na coluna primeiroNome
df_produto.drop_duplicates(subset=['primeiroNome'])
```

```python
# Deletando todos os produtos com o mesmo nome e o mesmo preço e considerando apenas o primeiro elemento encontrado
df_produto.drop_duplicates(subset=['primeiroNome', 'vlPreco'], keep='first')
```

# Agregações

```python
# Sumarizando os dados numéricos do dataframe utilizando o .describe()
df_produto['vlPrecoInflacao'] = (df_produto['vlPreco'] * 1.09).round(2)
df_produto.describe()
```

```python
# Utilizando .describe() para variáveis categóricas
df_produto['descItem'].describe()
```

```python
# Fazendo uma comparação dos dados da coluna descItem com uma coluna somente com os primeiros nomes de cada item
df_produto['descItemPrimeiro'] = df_produto['descItem'].apply(lambda x: x.lower().split(" ")[0])
df_produto[['descItem', 'descItemPrimeiro']].describe()
```

```python
# Fazendo uma tabela de frequência para as variáveis categóricas da coluna descItemPrimeiro utilizando uma função direta do pandas pd.value_counts()
freq_itens = pd.value_counts(df_produto['descItemPrimeiro'])
```

```python
# Fazendo um gráfico de frequência com os 5 primeiros itens de descItemPrimeiro
plt.bar(freq_itens.index[:5], freq_itens.values[:5])
# Realizando algumas configurações de DataVisualization
plt.grid(True)
plt.title("Top 5 Produtos")
plt.xlabel("Produtos")
plt.ylabel("Frequência")
plt.xticks(rotation=45)
```

## Group By

```python
# Agrupando os dados utilizando o .groupby() de acordo com os dados da tabela descItemPrimeiro e realizando a média para os valores da coluna vlPreco
df_produto.groupby(by=['descItemPrimeiro'])['vlPreco'].mean()
```

```python
# Utilizando o .describe() para os dados agrupados de descItemPrimeiro com relação a coluna vlPreco
df_produto.groupby(by=['descItemPrimeiro'])['vlPreco'].describe()
```

```python
# Passando uma lista de estatisticas que serão aplicadas as colunas desejadas utilizando o .agg()
df_produto.groupby(by=['descItemPrimeiro'])[['vlPreco', 'vlPrecoInflacao']].agg(['min', 'mean', 'max'])
```

```python
# Criando uma função que retorna um valor aleatorio de uma determinada serie
import random

def escolhe_aleatorio(x):
  return random.choice(x.values)
```

```python
# Utilizando a função escolhe_aleatorio como um parametro para o .agg()
(df_produto.groupby(by=['descItemPrimeiro'])[['vlPreco', 'vlPrecoInflacao']]
           .agg(['min', 'mean', 'max', escolhe_aleatorio]))
```

```python
# Utilizando a mesma função escolhe_aleatorio como um parametro para o .agg() com lambda function
(df_produto.groupby(by=['descItemPrimeiro'])[['vlPreco', 'vlPrecoInflacao']]
           .agg(['min', 'mean', 'max', lambda x: random.choice(x.values)]))
```
