# Merge

# Concat

```python
# Criando um df como exemplo
df_frutas = pd.DataFrame({
    'nome': ['Banana', 'Melancia', 'Morango', 'Mamao', 'Laranja'],
    'peso': [200, 800, 150, 420, 250]
})
df_frutas
```

```python
# Criando mais um df como exemplo
df_frutas_citricas = pd.DataFrame({
    'nome': ['Laranja', 'Limao', 'Mexirica', 'Limao Ciciliano'],
    'peso': [250, 240, 235, 200],
    'validade': [14, 16, 10, 12]
})
df_frutas_citricas
```

```python
# Utilizando o .concat() para concatenar os dataframes
pd.concat([df_frutas, df_frutas_citricas])
```

```python
# Utilizando o .sort_values() para ordenar o dataframe com relação ao nome e a validade e depois utilizando o .drop_duplicates() para eliminar as duplicatas
# depois utilizando o .fillna(0) para preencher os valores nan com 0
(pd.concat([df_frutas, df_frutas_citricas])
    .sort_values(['nome', 'validade'])
    .drop_duplicates(subset=['nome'])
    .fillna(0))
```

##  1 - Podemos remover algum tipo de massa do cardápio? Qual o impacto dessa ação?

```python
# Criando um filtro para filtrar as massas do df_item_pedido e atribuindo a um novo dataframe
filtro_massa = df_item_pedido['descTipoItem'] == 'massa'
df_massa = df_item_pedido[filtro_massa]
df_massa
```

```python
# Agrupando e contando todas as massas do df_item_pedido e utilizando o .reset_index() para transformar os index em colunas
df_group = (df_massa.groupby(by=['descItem'])['idPedido']
                    .nunique()
                    .reset_index())
df_group
```

```python
# Realizando uma comparação de porcentagem do total de cada uma das massas
df_group['Representa(%)'] = df_group['idPedido'] / df_group['idPedido'].sum()
df_group
```

Olhando esse resultado, podemos ver que uma resposta para a pergunta de qual massa podemos remover do cardápio tentando manter o mínimo de impacto possível, em relação ao número de pedidos a **massa integral** seria uma opção

```python
# Olhando esta questão com relação ao custo dos produtos, adicionando a tabela produto para comparações
df_massa = df_item_pedido[filtro_massa].merge(df_produto, how='left')
df_massa
```

```python
# Criando um df agrupando as massas com relação a quantidade de pedidos de cada uma das massas e a soma dos calores de cada massa com relação a quantidade de pedidos
df_group = (df_massa.groupby(by=['descItem'])
                    .agg({'idPedido': ['nunique'],
                          'vlPreco': ['sum']})
                    .reset_index())
df_group
```

```python
# Organizando a visualização das colunas
df_group.columns = df_group.columns.get_level_values(0)
```

```python
# Realizando uma comparação de porcentagem do total de cada uma das massas com relação a quantidade e aos seus preços
df_group['RepQtde(%)'] = (df_group['idPedido'] / df_group['idPedido'].sum()).round(2)
df_group['RepPreco(%)'] = (df_group['vlPreco'] / df_group['vlPreco'].sum()).round(2)
df_group
```

Quando analizado a representação do preço de custo da massa integral ela tem uma representatividade de 5% da receita o que corrobora com decisão de sua exclusão


## 2 - Podemos deixar a de Catupiry como borda padrão?

```python
# Criando um filtro para pegar somente as bordas do descTipoItem
filtro_borda = df_item_pedido['descTipoItem'] == 'borda'
```

```python
# Criando um novo dataframe contendo apenas as bordas
df_borda = df_item_pedido[filtro_borda]
df_borda
```

```python
# Criando um dataframe contendo os valores agrupados das quantidades de cada uma das bordas
df_group = (df_borda.groupby(['descItem'])['idPedido']
                    .nunique()
                    .reset_index())
df_group
```

```python
# Fazendo a representação percentual de cada uma das massas
df_group['RepPedidos(%)'] = (df_group['idPedido'] / df_group['idPedido'].sum()).round(2)
df_group
```

Vemos que a borda sabor catupiry representa cerca de 37% dos pedidos, perdendo para borda tradicional com 40% em relação ao total de pedidos. Podemos então inferir que não seria uma boa decisão deixar a borda de catupiry ser alocada como padrâo.

## 3 - Vários queijos estão estragando pela validade. Quais queijos juntos atendem 90% dos pedidos?

```python
# Vereficando quais queijos existem
df_item_pedido['descTipoItem'].unique()
```

```python
# Selecionando os dois tipos de queijo presente no dataframe e criando um filtro para eles
filtro_queijo = df_item_pedido['descTipoItem'].isin(['queijo 1', 'queijo 2'])
```

```python
# Criando um dataframe contendo apenas os queijos utilizando o filtro_queijo
df_queijo = df_item_pedido[filtro_queijo]
df_queijo
```

```python
# Agrupando os quejos pela quantidade de cada um deles nos pedidos e ordenando-os do maior para o menor
df_group = (df_queijo.groupby(['descItem'])['idPedido']
                     .nunique()
                     .reset_index()
                     .sort_values('idPedido', ascending=False))
df_group
```

```python
# Criando uma coluna com o percentual de vendas de cada um dos queijos
df_group['RepQueijo(%)'] = (df_group['idPedido'] / df_group['idPedido'].sum()).round(2)
df_group
```

```python
# Criando uma coluna com o percentual acumulado de cada um dos queijos
df_group['RepQueijo(%)Acum'] = df_group['RepQueijo(%)'].cumsum()
df_group
```

Analisando o resultado temos que até o queijo cheddar temos 92% dos pedidos que incluem queijo. Logo, tanto o queijo brie, queijo coalho, ricota e queijo feta estão fora dos 90% dos pedidos e podem ser considerados para o discarte.

## 4 - Quais ingredientes podemos remover do cardápio para diminuir custo no estoque?

```python
# Vereficando quais queijos existem
df_item_pedido['descItem'].unique()
```

```python
# Agrupando os ingredientes pela quantidade de cada um deles nos pedidos e ordenando-os do maior para o menor
df_group = (df_item_pedido.groupby(['descItem'])['idPedido']
                     .nunique()
                     .reset_index()
                     .sort_values('idPedido', ascending=False))
df_group
```

```python
# Criando uma coluna com o percentual de vendas de cada um dos ingredientes
df_group['RepIngredientes(%)'] = (df_group['idPedido'] / df_group['idPedido'].sum()).round(2)
df_group
```

```python
# Criando um filtro de ingredientes que foram pedidos menos de 10 vezes
filtro_ingredientes = df_group['idPedido'] <= 10
```

```python
# Buscando os ingredientes que foram pedidos menos de 10 vezes
df_group[filtro_ingredientes]
```

## 5 - Hoje só estamos em SP, quais os próximos 5 estados devemos abrir filiais?

```python
# Criando um dataframe que não inclua os pedidos do estado de São Paulo
df_not_sp = df_pedido[df_pedido['descUF']!='São Paulo']
```

```python
# Criando um dataframe com os pedidos agrupados por estados e ordenando-os do maior para o menor e selecionando os 5 primeiros
df_pedidos_estados = (df_not_sp.groupby('descUF')['idPedido']
                               .nunique()
                               .reset_index()
                               .sort_values('idPedido', ascending=False)
                               .head(5))
df_pedidos_estados
```

Temos que os 5 estados que mais tem pedidos depois de São Paulo são Rio de Janeiro, Minas Gerais, Paraná, Rio Grande do Sul e Pernambuco, logo, os 5 próximos estados ideais para a construção de novas filiais são eles.

## 6 - A preferência do tipo de massa muda entre os estados? E os ingredientes prediletos?

```python
# Criando um dataframe utilizando o filtro de massas e dando merge com a tabela de pedidos
df_massa = df_item_pedido[filtro_massa].merge(df_pedido[['idPedido', 'descUF']], how='left')
df_massa
```

```python
# Criando um dataframe contendo um agrupamento de pedidos de massa por estado, deixando-os ordenados pela estado e criando uma tabela pivot dos dados
df_massa_estado = (df_massa.groupby(['descItem', 'descUF'])['idPedido']
                           .nunique()
                           .reset_index()
                           .sort_values(['descUF', 'descItem'])
                           .pivot_table(values='idPedido',columns='descItem', index='descUF')
                           .reset_index()
                           .fillna(0))
df_massa_estado
```
