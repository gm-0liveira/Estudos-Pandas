```python
import pandas as pd
import numpy as np
```

# Transformações

```python
df = pd.read_csv("/content/drive/MyDrive/Estudos/Desbravando Pandas/data/pedido.csv")
df
```

```python
# Renomeando Colunas
df = df.rename(columns= {"descUF": "descEstado"})
```

```python
# Renomeando Colunas com inplace
df.rename(columns= {"descEstado": "descUF"}, inplace=True)
```

```python
# Inicializando o Dataframe de Produtos
df_produtos = pd.read_csv('/content/drive/MyDrive/Estudos/Desbravando Pandas/data/produto.csv')
df_produtos
```

```python
# Criando uma nova coluna , com os preços dos produtos ajustados a 9%
df_produtos['precoAjustado'] = df_produtos['vlPreco'] * 1.09
df_produtos
```

```python
# Arredondando as casas decimais de uma coluna para 2 casas
df_produtos['precoAjustado'] = df_produtos['precoAjustado'].round(2)
df_produtos
```

```python
# Unindo a criação de uma nova coluna com o arredondamento dos valores
df_produtos['precoAjustado'] = (df_produtos['vlPreco'] * 1.09).round(2)
df_produtos
```

```python
# Calculando a taxa adicionada a cada linha e adicionando seu valor arredondado em uma nova coluna
df_produtos['txAjuste(%)'] = (100 * (df_produtos['precoAjustado'] / df_produtos['vlPreco'] - 1)).round(2)
df_produtos
```

```python
# Deletando uma coluna do dataframe
df_produtos = df_produtos.drop(columns=['txAjuste'])
```

```python
# Aplicando log a cada um dos valores da coluna txAjuste e salvando em outra coluna através do numpy
df_produtos['logTxAjuste'] = np.log(df_produtos['txAjuste(%)'])
df_produtos
```

```python
# Criando uma função (exemplo) que realiza um fatorial para um número
def fatorial(x):
  total=1
  for i in range(2,int(x)+1):
    total *= i
  return total
```

```python
# Aplicando a função fatorial utilizando o .apply() para cada um dos elementos da coluna precoAjustado
df_produtos['precoAjustado'].apply(fatorial)
```

# Filtros

```python
# Filtrando apenas clientes de São Paulo
df[df['descUF'] == 'São Paulo']
```

```python
# Filtrando dados de acordo com os valores de duas colunas, é de São Paulo e quer ketchup
filtro_sp_ketchup = (df['descUF'] == 'São Paulo') & (df['flKetchup'])
df[filtro_sp_ketchup]
```

```python
# Filtrando Pedidos de São Paulo, Rio de Janeiro, Parana e que pediram ketchup
filtro_sp_pr_rio_ketchup = ((df['descUF'] == 'São Paulo') | (df['descUF'] == 'Rio de Janeiro')
| (df['descUF'] == 'Paraná')) & (df['flKetchup'])
df[filtro_sp_pr_rio_ketchup]
```

```python
# Verificando os valores unicos de uma série
df[filtro_sp_pr_rio_ketchup]['descUF'].unique()
```

```python
# Filtrando Pedidos de São Paulo, Rio de Janeiro, Parana e que pediram ketchup utilizando o .isin()
sp_pr_rio = ['São Paulo', 'Rio de Janeiro', 'Paraná']
filtro_sp_pr_rio_ketchup = df['descUF'].isin(sp_pr_rio) & df['flKetchup']
df[filtro_sp_pr_rio_ketchup]
```
